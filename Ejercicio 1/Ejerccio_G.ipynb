Deseamos saber qué directores han dirigido el máximo número de películas, junto con ese número de películas.

````python
# Esta celda debe ser completada por el estudiante
def directors_max_movies_df(df):
    sol = dict(df["director_name"].value_counts())
    sol.pop("Desc")

    max_movies = 0
    for key, value in sol.items():
        if value > max_movies and key:
            sol=[key]
            max_movies = value
        elif value == max_movies:
            sol.append(key)
        else:
            # Añado este break porque se que el metodo .value_counts devuelve un objeto ordenado 
            break
    
    return sol, max_movies
````

````python
def gini_float(y):
    n = len(y)
    if n == 0:
        return 0
    fr = [y.count(i) for i in set(y)]
    return sum([(fr[i] / n) * (n - fr[i]) for i in range(len(fr))])
    
def gini_bool(y):
    n = len(y)
    if n == 0:
        return 0
    return 1 - (y.count(True)/n)**2 - (y.count(False)/n)**2


def mode_max(y):
    if len(y) == 0:
        return -np.inf
    count = np.bincount(y)
    return np.argmax(count)

def branch(x, y, S, col_types, col_usadas):
    best_score = np.inf
    for j in range(len(x[0])):
        # print(col_usadas)
        if j in col_usadas:
            continue
        if col_types[j] == bool:
            f = gini_bool
        elif col_types[j] == float:
            f = gini_float
        #print(f"cambiamos a la columna {j}")
        left=[]
        right=[]
        y_left=[]
        y_right=[]
        for i in S:
            if col_types[j] == bool:
                    #print(f"vamos a ver este valor {x[i][j]}")
                if x[i][j]:
                    left.append(i)
                    y_left.append(y[i])
                else:
                    right.append(i)  
                    y_right.append(y[i])

                left_score = f(y_left)
                right_score = f(y_right)
                score = left_score + right_score
            elif col_types[j] == float:
                left = []
                right = []
                y_left = []
                y_right = []
                for k in S:
                    #print(f"vamos a ver este valor {x[k][j]}")
                    if x[k][j] < x[i][j]:
                        left.append(k)
                        y_left.append(y[k])
                    else:
                        right.append(k)
                        y_right.append(y[k])
                
                left_score = f(y_left)
                right_score = f(y_right)
                score = left_score + right_score
            #print(f"nuestros scores son:{score}, left {left_score}, right {right_score}")
            if score < best_score:
                best_score = score
                i_1 = i
                j_1 = j
                left_1 = left
                right_1 = right
                left_score_1 = left_score
                right_score_1 = right_score    

    #print(f"En este nodo nos quedamos con un {left_1}, {right_1}, y asu vez tenemos un score de {score}, para la columna {j_1}")
    return [i_1, j_1, left_1, right_1, best_score, left_score_1, right_score_1]


class Stack:
    def __init__(self,parent,indexes,score,col_usadas=[]):
        self.parent=parent
        self.indexes=indexes
        self.score=score
        self.col_usadas = col_usadas


class Node:
    def __init__(self,parent,resultado,col_quest,th,indexes):
        self.parent=parent
        self.resultado=resultado
        self.col_quest=int(col_quest)
        self.th=th
        self.indexes=indexes
        self.left = None
        self.right = None

    def ask_question(self, new_row, node, indice, col_types, col_names):
        print(f"Nos encontramos en el nodo de pregunta numero {indice}")
        if self.resultado:
            print("Creo que me puedo imaginar que la solucion es:")
            return bool(self.center)
        
        if col_types[self.col_quest] == int:
            print(f"¿Es mayor el {col_names[self.col_quest]} que {self.th}?")
            if new_row[self.col_quest]>self.th:
                print("Si")
                return node[self.right].ask_question(new_row, node, self.right, col_types, col_names)
            else:
                print("No")
                return node[self.left].ask_question(new_row, node, self.left, col_types, col_names)

        elif col_types[self.col_quest] == bool:
            print(f"¿Es del genero {col_names[self.col_quest]}?")
            if new_row[self.col_quest]:
                print("Si")
                return node[self.right].ask_question(new_row, node, self.right, col_types, col_names)
            else:
                print("No")
                return node[self.left].ask_question(new_row, node, self.left, col_types, col_names)


def dt(x, y, col_types, alpha=0, n_min=1):
        # A single set of stack is constructed. Decision tree is initialized.
    stack = [Stack(0, list(range(len(x))), gini_bool(y))]  # f is global
    node = []
    k = -1

    while len(stack) > 0:
        popped = stack.pop()
        k = k + 1

        i, j, left, right, score, left_score, right_score = branch(x, y, popped.indexes, col_types, popped.col_usadas)
        col_usadas = popped.col_usadas+[j]
        if abs(popped.score - score) < alpha or len(popped.indexes) < n_min or len(left) == 0 or len(right) == 0 or len(col_usadas)==len(x[0]):
            node.append(Node(popped.parent, True, j, 0, popped.indexes))
            #print(f"El problema que nos esta dando viene de {popped.score - score < alpha},{len(popped.indexes) < n_min} ,{len(left) == 0 or len(right) == 0}")
            #print(f"nuestro score es {score} y nuestro popped score es {popped.score} por lo que nuestra suma se qued acomo {abs(popped.score - score)}")
        else:
            node.append(Node(popped.parent, False, j, x[i][j], popped.indexes))
            stack.append(Stack(k, right, right_score,col_usadas))
            stack.append(Stack(k, left, left_score,col_usadas))


    for h in range(k, 0, -1):
        pa = node[h].parent
        if node[pa].right == None:
            node[pa].right = h
        elif node[pa].left == None:
            node[pa].left = h
        
    for h in range(k + 1):
        if node[h].resultado:
            y_node = [y[index] for index in node[h].indexes]
            #print(f"el nodoy es {y_node}")
            node[h].center = mode_max(y_node)
        else:
            node[h].center = None

    return node



nodes = dt(x, y, col_types, n_min=2)
print(nodes)
new_row = [True, True, False, 185000000.0]
print(nodes[0].ask_question(new_row, nodes, 0, col_types, col_name))
print(f"la longiud de nodos es:{len(nodes)}")
for node in nodes:
    print(f"el padre es {node.parent}, mientras que la j es {node.col_quest}")
    print(node.center)
````
